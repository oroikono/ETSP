# 1. Load GraphSay and create the instance.

```python
from GraphSay.CaptioningAgent import Pix2Struct_Base

# Initialize GraphSay with the name of the pre-trained model you want to use
graph_say = Pix2Struct_Base()
```

# 2. Generate Captions

To maximize the accuracy of generated captions, try to follow the guideline below when you plotting.

__Single column data__


In the case that you want to plot a single column data which has one Y value for each X value, double check whether your plot has a correct <ins>axis lable</ins> and <ins>title</ins>!


__Multi column data__


In the case that you want to plot a multi column data which has multiple Y value for each X value, double check whether your plot not only has a correct <ins>axis lable</ins> and <ins>title</ins>, but also plotting in <ins>one plot together</ins> with a correct <ins>legends</ins>!

## a. Combined with Matplotlib

```python
import pandas as pd

# Sample data
sample_data = pd.read_csv("./GraphSay/sample_dataset.csv",index_col=0)
xlabel = 'Country'
xticks = sample_data.index
ylabel = 'Share of players'
title = 'Distribution of players of PlayerUnknown in Battlegrounds ( PUBG ) worldwide as of June 2017 , by country'

sample_data.head()
```

```python
ax = sample_data.plot.bar(rot=0)
  
_ = ax.set_xlabel(xlabel) 
_ = ax.set_ylabel(ylabel) 
_ = ax.set_title(title)
_ = ax.set_xticklabels(xticks, rotation = 30)

matplotlib_fig = ax.get_figure()

# Generate captions for the matplotlib plot
caption_matplotlib = graph_say.generate_caption(matplotlib_fig)
print(f"Generated caption for matplotlib plot:\n{caption_matplotlib}")
```

## b. Combined with Seaborn

```python
import pandas as pd

# Sample data
sample_data = pd.read_csv("./GraphSay/sample_dataset.csv",index_col=0)
xlabel = 'Country'
xticks = sample_data.index
ylabel = 'Share of players'
title = 'Distribution of players of PlayerUnknown in Battlegrounds ( PUBG ) worldwide as of June 2017 , by country'

sample_data.reset_index(inplace=True)

sample_data.head()
```

```python
import seaborn as sns

ax = sns.barplot(data=sample_data, x=sample_data["Response"], y=sample_data["Share of players"])
_ = ax.set(xlabel=xlabel, ylabel=ylabel, title=title)
_ = ax.tick_params(axis='x', rotation=30)

sns_fig = ax.get_figure()

# Generate captions for the seaborn plot
caption_seaborn = graph_say.generate_caption(sns_fig)
print(f"Generated caption for matplotlib plot:\n{caption_seaborn}")
```
